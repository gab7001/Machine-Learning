# import libraries

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generate artificial dataset with make_blobs. I'll use 300 samples, 2 features, 4 centers

```python
X,y=make_blobs(n_samples=300, n_features=2, centers=4, cluster_std=4, random_state=10)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(X[:,0],X[:,1]) #paint the first and second features
```

## set algorithm

```python
model = KMeans(n_clusters=4) #call the algorithm with 4 clusters and assign it to the variable model
model.fit(X) #fit the model to the data
```

## Predict

```python
model_predict = model.predict(X) #execute the model and assign to a variable
centroids = model.cluster_centers_ #identify the centroids
print(model.cluster_centers_)
```

## Visualize

```python
plt.figure(figsize=(7,5))
plt.scatter(X[:,0],X[:,1],c=model_predict, s=50, cmap='rainbow')
plt.scatter(centroids[:,0],centroids[:,1],c='black',s=200,alpha=1)
```

## Another model with 6 clusters

```python
model1=KMeans(n_clusters=6)
model1.fit(X)
```

```python
model1_predict = model1.predict(X)
centroids1 = model1.cluster_centers_
print(model1.cluster_centers_)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(X[:,0],X[:,1],c=model1_predict,s=50,cmap='rainbow')
plt.scatter(centroids1[:,0],centroids1[:,1],c='black',s=200,alpha=1)
```
